```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_grad(x):
    return (x > 0).astype(float)

x = np.linspace(-10, 10, 500)
y = relu(x)
dy = relu_grad(x)

# Wykres
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="ReLU", linewidth=2)
plt.plot(x, dy, label="ReLU - gradient", linestyle='--')
plt.title("Funkcja ReLU i jej gradient")
plt.xlabel("x")
plt.ylabel("wartość")
plt.grid(True)
plt.legend()
plt.show()
```
